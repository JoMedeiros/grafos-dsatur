## Função para verificar compatibilidade de horários

```python
''' Compara se dois horários tem mesmo turno '''
def comp_t(hor1, hor2):
    turno = ['M','T','N']
    comp_turno = False
    for t in turno:
        h1 = hor1.find(t)
        h2 = hor2.find(t)
        if (h1 > 0 and h2 > 0):
            return t
    return 'F'

''' Compara se há confilto entre horários '''
def comp(hor1, hor2):
    turno = comp_t(hor1, hor2)
    if (turno == 'F'):
        return False
    
    w1, h1 = hor1.split(turno)
    w2, h2 = hor2.split(turno)
    
    for d1 in w1:
        if (d1 in w2):
            for h in h1:
                if(h in h2):
                    return True
    return False
```

```python
g = nx.Graph()
#for i in range()
#for h1 in turmas['Horário']:
#    for h2 in turmas['Horário']
n = len(turmas['Horário'])
for i in range(n):
    for j in range(i, n):
        if (comp(turmas['Horário'][i], turmas['Horário'][j])):
            g.add_edge(i, j)
```

```python
# Com problemas hahaha
def DSATUR(G):
    next_color = 2
    colors = {}
    V = [True for key in G.nodes]
    degrees = [G.degree(key) for key in G.nodes]
    dsat = [0 for key in G.nodes]
    
    g = get_great_grade(G, V, degrees)
    V[g] = False
    update_degrees(G, degrees, g)
    
    colors[g] = 1
    for i in list(G.neighbors(g)):
        dsat[i] = 1
    
    count = len(G.nodes) -1
    while(count > 0):
        sat = 0
        dg = 0
        n = -1
        for i in G.nodes:
            if(V[i]):
                if(dsat[i] > sat):
                    n = i
                    sat = dsat[i]
                    dg = degrees[i]
                else:
                    if(dsat[i] == sat and degrees[i] > dg):
                        n = i
                        sat = dsat[i]
                        dg = degrees[i]
                        
        n_c = get_next_color(G, n, colors)
        if (n_c == 1):
            colors[n] = next_color
            next_color = next_color + 1
        else:
            if(n_c == 1000):
                colors[n] = 1
            else:
                colors[n] = n_c - 1
            
        V[n] = False
        update_degrees(G, degrees, n)
        for i in list(G.neighbors(n)):
            update_dsat(G, dsat, colors, i)
        
        count = count - 1
    
    return colors
```
